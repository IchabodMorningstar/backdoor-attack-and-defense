## compare two images using histrogram comparison-HSV color space

```python
import cv2
# Load the images
img1 = cv2.imread('gtsrb-german-traffic-sign/Patches/patch_4.png')
img2 = cv2.imread('gtsrb-german-traffic-sign/Train_official/14/00014_00000_00001.png')

# Convert it to HSV
img1_hsv = cv2.cvtColor(img1, cv2.COLOR_BGR2HSV)
img2_hsv = cv2.cvtColor(img2, cv2.COLOR_BGR2HSV)

# Calculate the histogram and normalize it
hist_img1 = cv2.calcHist([img1_hsv], [0,1], None, [180,256], [0,180,0,256])
cv2.normalize(hist_img1, hist_img1, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX);
hist_img2 = cv2.calcHist([img2_hsv], [0,1], None, [180,256], [0,180,0,256])
cv2.normalize(hist_img2, hist_img2, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX);

# find the metric value
metric_val = cv2.compareHist(hist_img1, hist_img2, cv2.HISTCMP_BHATTACHARYYA)
print(metric_val)
```

## compare two images using histrogram comparison-YCrCb color space

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def compare_images_with_visualization(img_path1, img_path2):
        # Load the images
    img1 = cv2.imread(img_path1)
    img2 = cv2.imread(img_path2)
    # Convert to YCrCb
    img1_ycrcb = cv2.cvtColor(img1, cv2.COLOR_BGR2YCrCb)
    img2_ycrcb = cv2.cvtColor(img2, cv2.COLOR_BGR2YCrCb)

    # Initialize an array to hold the comparison results
    comparison_results = []

    # Channel names for visualization
    channels = ['Y', 'Cr', 'Cb']

    # Plotting setup
    fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 8))
    fig.suptitle('Histogram Comparison of YCrCb Channels')

    # Compare each channel (Y, Cr, Cb)
    for i in range(3):
        hist_img1 = cv2.calcHist([img1_ycrcb], [i], None, [256], [0, 256])
        hist_img2 = cv2.calcHist([img2_ycrcb], [i], None, [256], [0, 256])

        cv2.normalize(hist_img1, hist_img1, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX)
        cv2.normalize(hist_img2, hist_img2, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX)

        # Plot histograms
        axes[i, 0].plot(hist_img1, color='blue')
        axes[i, 0].set_title(f'Image 1 - {channels[i]} Channel')
        axes[i, 1].plot(hist_img2, color='red')
        axes[i, 1].set_title(f'Image 2 - {channels[i]} Channel')

        # Compare using Bhattacharyya distance
        comparison_results.append(cv2.compareHist(hist_img1, hist_img2, cv2.HISTCMP_BHATTACHARYYA))

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    #plt.show()

    # Aggregate results (mean for simplicity)
    overall_comparison = np.mean(comparison_results)
    return overall_comparison


image_path1 = r"D:\IchMorningstar\Research\MLLsecurity\gtsrb-german-traffic-sign\patch\patch1.png"
image_path2 = r"D:\IchMorningstar\Research\MLLsecurity\gtsrb-german-traffic-sign\Backdoored_test\clean\00014_00033_00006.png"
comparison_score = compare_images_with_visualization(image_path1, image_path2)
print(comparison_score)
# starting_batch=0
# for image_id in range(0,100 ):
#     # Example usage
#     batch_id = starting_batch + (image_id // 30)
#     img_id_in_batch = image_id % 30
#     image_path1 = 'gtsrb-german-traffic-sign/Patches/rock1.jpeg'
#     image_path2 = 'gtsrb-german-traffic-sign/Patches/rock2.jpeg'
#     f"00014_{batch_id:05}_{img_id_in_batch:05}.png"
#     comparison_score = compare_images_with_visualization(image_path1, image_path2)
#     print(f"Image {image_path1} Comparison Score:", comparison_score)
```

## compare two images using SSIM method-RGB color space with custom position of patch

```python
import cv2
from skimage.metrics import structural_similarity as ssim
import matplotlib.pyplot as plt
import numpy as np

def crop_patch(image, top_left, patch_size):
    # Ensure that we do not go beyond the image boundaries
    if top_left[0] + patch_size[0] > image.shape[1] or top_left[1] + patch_size[1] > image.shape[0]:
        return None
    return image[top_left[1]:top_left[1] + patch_size[1], top_left[0]:top_left[0] + patch_size[0]]

patch_path = [r"D:\IchMorningstar\Research\MLLsecurity\gtsrb-german-traffic-sign\patch\patch1.png",
                r"D:\IchMorningstar\Research\MLLsecurity\gtsrb-german-traffic-sign\patch\patch2.png",
                r"D:\IchMorningstar\Research\MLLsecurity\gtsrb-german-traffic-sign\patch\patch3.png",
                r"D:\IchMorningstar\Research\MLLsecurity\gtsrb-german-traffic-sign\patch\patch4.png",
                r"D:\IchMorningstar\Research\MLLsecurity\gtsrb-german-traffic-sign\patch\patch5.png",
                r"D:\IchMorningstar\Research\MLLsecurity\gtsrb-german-traffic-sign\patch\patch6.png",
                r"D:\IchMorningstar\Research\MLLsecurity\gtsrb-german-traffic-sign\patch\blue.png",
                r"D:\IchMorningstar\Research\MLLsecurity\gtsrb-german-traffic-sign\patch\red.png"
             ]
image_path1 = patch_path[6]
image_path2 = r"D:\IchMorningstar\Research\MLLsecurity\gtsrb-german-traffic-sign\Backdoored_test\clean\00014_00033_00006.png"
# Load images in BGR color space (OpenCV's default)
image1_bgr = cv2.imread(image_path1)
image2_bgr = cv2.imread(image_path2)
image1_bgr = cv2.resize(image1_bgr, (30, 30))
image2_bgr = cv2.resize(image2_bgr, (256, 256))

# Define the size of the filter
filter_size = (30, 30)  # Example size, adjust to your needs
# filter_patch = crop_patch(image2_bgr, (0, 0), filter_size)

# Display the filter patch (convert to RGB for display)
plt.imshow(cv2.cvtColor(image1_bgr, cv2.COLOR_BGR2RGB))
plt.title("Filter Patch")
plt.show()

# Define the step size for sliding the window
step = 20  # Adjust the step size as needed

# Create a copy of the second image for visualization
visual_image = image2_bgr.copy()
index_lst = []
# Slide the window over the image
for y in range(0, image2_bgr.shape[0] - filter_size[1], step):
    for x in range(0, image2_bgr.shape[1] - filter_size[0], step):
        patch_2 = crop_patch(image2_bgr, (x, y), filter_size)

        if patch_2 is not None:
            ssim_index = ssim(image1_bgr, patch_2, full=False, multichannel=True, win_size=3)
            index_lst.append(ssim_index)

            if ssim_index > 0:
                # Highlight the current window position
                cv2.rectangle(visual_image, (x, y), (x + filter_size[0], y + filter_size[1]), (255, 0, 0), 2)

                # Display the image with the highlighted window
                plt.imshow(cv2.cvtColor(visual_image, cv2.COLOR_BGR2RGB))
                plt.title(f"Window at ({x}, {y}), SSIM: {ssim_index:.2f}")
                plt.show()
                plt.pause(0.1)  # Pause to create an animation effect
                plt.clf()

                # Reset the visual image for the next iteration
                visual_image = cv2.cvtColor(image2_bgr, cv2.COLOR_BGR2RGB)

# You can adjust the 'filter_size' and 'step' variables as needed for your specific images and tasks.
```
